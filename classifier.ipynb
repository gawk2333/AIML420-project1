```python
#load data
import numpy as np
import pandas as pd
from scipy.io.arff import loadarff

data1, meta1 = loadarff('./data/banknotes.arff')
data2, meta2 = loadarff('./data/ionosphere.arff')
data3, meta3 = loadarff('./data/steel-plates-fault.arff')

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df3 = pd.DataFrame(data3)
```

```python
#split data
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

def prep_data(df):
    # Separate features (X) and target variable (y)
    x = df.drop(df.columns[-1],axis=1)
    y = df[df.columns[-1]]
    le = LabelEncoder()
    y = le.fit_transform(y)
    # Use train_test_split to split the data into training and testing sets
    train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2, random_state=42)

    return train_x,test_x,train_y,test_y
```

```python
#KNN, setting k - "n_neighbors" 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

def knn_classifier(df):
    map = {}
    for k in range(1,50):
        train_x,test_x,train_y,test_y = prep_data(df)
        neigh = KNeighborsClassifier(n_neighbors = k)
        neigh.fit(train_x,train_y)
        predictions = neigh.predict(test_x)
        # print(f"typeofpre:{type(predictions)},typeoftest_x:{type(test_x)}")
        accuracy = accuracy_score(test_y, predictions)
        map[k] = accuracy
    return map

m1 = knn_classifier(df1)
m2 = knn_classifier(df1)
m3 = knn_classifier(df1)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
#function to find the best k
def best_k(df,n_range,target_y):
    #knn for steel-plates-fault
    x1 = df.drop(target_y, axis=1)
    y1 = df[target_y]
    # Encode the labels
    le = LabelEncoder()
    y1 = le.fit_transform(y1)
    #n_neighbors 1 to 10
    n_neighbors = range(1, n_range)
    # save accuracies and n_neighbors as a dictionary
    acc = {}
    # Loop over different values of k
    for i in n_neighbors:
        X_train, X_test, y_train, y_test = train_test_split(x1, y1, test_size=0.2, random_state=42)
        knn = KNeighborsClassifier(n_neighbors=i)
        knn.fit(X_train, y_train)
        y_pred = knn.predict(X_test)
        #add accuracy_score and n_neighbors
        acc[i] = accuracy_score(y_test, y_pred)
    return acc
```
